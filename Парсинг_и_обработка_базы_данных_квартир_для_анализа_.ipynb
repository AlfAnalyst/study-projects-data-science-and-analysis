###Задача:

Взять базу квартир (moscow.csv), провести парсинг и очистку данных:

● Убрать строки “Для заметок”;

● Убрать столбцы “ГРМ” и “Примечания”;

● Для столбца “Балкон” добавить категорию “Отсутствует” (вместо NaN);

● Значения столбца “Бонус агенту” перевести в тысячи рублей,а пустые значения
заполнить нулями;

● В столбец “Площадь” записать общую площадь всех помещений;

● Для столбца “Санузел” заполнить отсутствующие значения строкой “Нет данных”;

● Столбец “Цена” перевести в тысячи рублей.

```
import pandas as pd             # Импортируем библиотеку обработки и анализа данных pandas
import matplotlib.pyplot as plt # Импортируем модуль pyplot библиотеки matplotlib для построения графиков
import datetime                 # Импортируем модуль datetime для обработки данных в формете даты
import numpy as np              # Импортируем библиотеку numpy
from google.colab import files  # Импортируем библиотеку files для работы с файлами
import re                       # Регулярные выражения

# Рисовать графики сразу же
%matplotlib inline
```

```
mos_df = pd.read_csv('/content/drive/MyDrive/Базы/moscow.csv', sep = ';') # Загружаем базу.
mos_df.head()
```

```
mos_df.shape # Размер загруженной базы.
```

```
mos_df.drop(['ГРМ', 'Примечание'], axis = 1, inplace = True) # Удаляем 2 столбца: "ГРМ" и "Примечание".
mos_df.tail(10)
```

```
mos_df.Балкон = mos_df['Балкон'].fillna('Отсутствует') #  Заменяем осмысленными значениями все пропуски.
mos_df.Санузел = mos_df['Санузел'].fillna('Нет данных')
mos_df['Бонус агенту'] = mos_df['Бонус агенту'].fillna(0)
mos_df[2000:2015]
```

```
mos_df_1 = mos_df.dropna(axis = 0, thresh = 5) # Удалим пустые строки, где непустых ячеек < 5. Это будут все строки "Для заметок".
mos_df_1.tail(20)
```

```
mos_df_1['От станции'] = mos_df_1['От станции'].fillna('Нет данных') # Заполним пустые ячейки в столбце "От станции".
mos_df_1.tail(10)
```

```
mos_df_2 = mos_df_1.copy() # Делаем копию базы.
```

```
# Функция будет обрабатывать столбец "Площадь".
def getParameterArea(arg):
  sarg=arg.split('/')
  item = sarg[0] # Берем только первый элемент, где содержится информация об общей площади.
  item = item.strip()
  return item
```

```
mos_df_2['Площадь'] = mos_df_1['Площадь'].apply(getParameterArea) # Применяем функцию.
mos_df_2.tail()
```

```
mos_df_2['Площадь'].unique()
```

```
mos_df_2['Бонус агенту'].unique() # Смотрим на содержимое столбца "Бонус агенту".
```

```
def getBonus(arg): # Функция будет обрабатывать столбец "Бонус агенту".
  for i in range(len(mos_df_2)):
    if 'руб' in str(arg):  #  Если есть 'руб'.
      global res
      res = 0
      arg = str(arg)
      res = int(re.sub(r'[^0-9]+', '', arg))   # Все, что не цифры, меняем на отсутствие символов.
      return int(res / 1000)
    if "%" in str(arg): # Если есть "%".
      res = mos_df_2['Цена, руб.'][i]*(float(re.sub(r'[^0-9]+', '', arg))/100)
      return int(res / 1000)
    else:
      return 0
```

```
mos_df_2['Бонус агенту'] = mos_df_2['Бонус агенту'].apply(getBonus) # Применяем функцию.
mos_df_2
```

```
mos_df_3 = mos_df_2.rename(columns={'Цена, руб.':'Цена (тыс.руб.)', 'Бонус агенту':'Бонус агенту (тыс.руб.)'}) # Меняем названия столбцов.
mos_df_3['Цена (тыс.руб.)'] = round(mos_df_2['Цена, руб.'] / 1000, 2) # Переводим значения в тыс.руб.
mos_df_3.head()
```

```
mos_df_3['Бонус агенту (тыс.руб.)'].unique() # Смотрим на новое содержимое столбца "Бонус агенту (тыс.руб.)".
```

```
mos_df_3.index = range(1, len(mos_df_3.index) + 1) # Обновляем индексный столбец.
mos_df_3.head()
```
